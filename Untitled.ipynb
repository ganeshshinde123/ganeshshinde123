# 3 types of higher order fun
###1. filter
###2.map
###3.reduce



```python
# filter : is used to filter out the values based on some condition
k = ['comp', 'entec','comp', 'entec','mech','comp','entec','comp', 'entec','mech']
print(k)
```

```python
#filter(function,sequence/iterable)
def find(dept):
    if dept == "entec":
        return True
    else:
        return False
find('entec')
```

```python
filter(find,k)
# whenever we get output in the form of object then typecast 
#that object or iterate the object 
# filter only work on true values
```

```python
list(filter(find,k))
```

```python
tuple(filter(find,k))
```

```python
# using lambda fun
filter(lambda dept:dept=='mech',k)
```

```python
list(filter(lambda dept:dept=='mech',k))
```

# map fun

```python
# map(function,sequence)
    # it is a fin which will be applied on each and every element of a sequence
n = lambda num:num + 100
n(95)
```

```python
d = [10,20,30,40]
map(n,d)
```

```python
list(map(n,d))
```

```python
# diff bet filter and map : filter is subset of original set  and map is same set of element in op
#similarities both gives obj op
```
